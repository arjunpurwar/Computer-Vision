```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.image as mpimg
import cv2
```

```python
image = mpimg.imread('download.jpg')

grey_image= cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(grey_image, cmap= 'gray')
```

```python
# Printling the value of particular pixel

x= 25
y= 70

print(grey_image[y,x])
```

```python
# Maximum pixel in image
print(np.amax(grey_image))

#Minimum pixel in image
print(np.amin(grey_image))
```

```python
#Shape of image
print(image.shape)

print(grey_image.shape)
```
