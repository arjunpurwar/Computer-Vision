```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image= cv2.imread('pizza_bluescreen.jpg')

print(image.shape)
type(image)
```

```python
plt.imshow(image) 
# The image is not right because the CV2 converts the images into BGR format. So, we now need to convert in into its orginal format i.e. RBG format.
```

```python
image_copy = np.copy(image)

image_copy= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
lower_blue= np.array([0,0,200])

upper_blue= np.array([250,250,255])
# Area of range to get rid off
```

```python
#Create the Mask- used to isolate the area of interest
mask= cv2.inRange(image_copy, lower_blue, upper_blue)

plt.imshow(mask, cmap= 'gray')
```

```python
masked_image = np.copy(image_copy)

masked_image[mask !=0]= [0,0,0]

plt.imshow(masked_image)
```

```python
background_image = cv2.imread('space_background.jpg')

background_image = cv2.cvtColor(background_image, cv2.COLOR_RGB2BGR)
```

```python
crop_background = background_image[0:514, 0:816]
```

```python
plt.imshow(crop_background)
```

```python
crop_background[mask == 0]= [0,0,0]
```
